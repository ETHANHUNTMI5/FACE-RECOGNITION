## Read a Video Stream from webCam

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame= cap.read()
    grayFRAME = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if ret== False:
        continue
    
    cv2.imshow("GRAY Frame",grayFRAME)
    cv2.imshow("Video Frame",frame)
    
    # Wait for user input -q , the you will stop
     # program will stop when user presses the q (as in quit ) button
    key_pressed = cv2.waitKey(1) & 0xFF
    
    if key_pressed== ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    res,frame = cap.read()
    
    if(res==False):
        continue
    
    grayF = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow("frame",frame)
    cv2.imshow("gframe",grayF)
    
    key_pressed = cv2.waitKey(1) & 0xff
    
    if(key_pressed==ord('q')):
        break
        
cap.release()
cv2.destroyAllWindows()
    
        
```
